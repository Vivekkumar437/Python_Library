```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips,color='green')
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=tips,color='blue')
plt.title("Line plot for total bill by size")
plt.show()
```

```python
##Bar plot

sns.barplot(x='day',y='total_bill',data=tips,color='red')
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.show
```

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
sns.pairplot(tips)
```

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,)
```
